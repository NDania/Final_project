#  <a name="5">5. Проверка гипотез </a>

H0: Средние пользовательские рейтинги на вино Испании и Португалии одинаковые.
H1:Средние пользовательские рейтинги на вино Испании и Португалии разные.
H0:Средние цены двух популярных сортов вина одинаковые.
H1:Средние цены двух популярных сортов вина разные.
Задать самостоятельно пороговое значение alpha.



```python
Ho = "Рейтинги на вино Испании и Португалии одинаковые"
Ha = "Рейтинги на вино Испании и Португалии разные"
t = df_two[df_two.country == 'Spain']
g = df_two[df_two.country == 'Portugal']
x = np.array(t[t.price < 300].points)
y = np.array(g[g.price < 300].points)
```

```python
import scipy.stats as stat
value, p = stat.normaltest(t[t.price < 400].points.sample(300))
print(value, p)
if p < 0.05:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

```python
results = stat.levene(x, y)

alpha = 0.05

print(results.pvalue)

if (results.pvalue < alpha):
    print('Отклонить гипотезу о равенстве дисперсий')
else:
    print('Принять гипотезу о равенстве дисперсий')
```

```python
t,p_value  = stat.ttest_ind(x,y, axis = 0, equal_var=False)

if p_value < 0.05:
    print(f'{Ha}, так как p-value - {p_value.round(3)} < 0.05')
else:
    print(f'{Ho}, так как p-value - {p_value.round(3)} > 0.05')
```

```python
t,p_value  = stat.ttest_ind(x,y, axis = 0, equal_var=False, alternative='less')

if p_value < 0.05:
    print(f'{Ha}, так как p-value - {p_value.round(3)} < 0.05')
else:
    print(f'{Ho}, так как p-value - {p_value.round(3)} > 0.05')
```

###Теперь проверим вторую пару гипотез.

```python
Ho = "Средние цены двух популярных сортов вина Испании и Португалии одинаковые"
Ha = "Средние цены двух популярных сортов вина Испании и Португалии разные"

t = df_spain[df_spain['variety'] == 'Tempranillo']
g = df_porty[df_porty['variety'] == 'Portuguese Red']
x = np.array(t[t.points > 95].price)
y = np.array(g[g.points > 98].price)
```

```python
from statsmodels.tools.tools import scipy
s, p = scipy.stats.shapiro(t['price'])
if p > 0.05:
    print('Pаспределено нормально' , p , s)
else:
     print('Pаспределено НЕ нормально', p, s)
```

```python
from statsmodels.tools.tools import scipy
s, p = scipy.stats.shapiro(g['price'])
if p > 0.05:
    print('Pаспределено нормально' , p , s)
else:
     print('Pаспределено НЕ нормально', p, s)
```

```python
l, pl = stat.levene(x, y)

alpha = 0.05

print(pl)

if (pl < alpha):
    print('Отклонить гипотезу о равенстве дисперсий')
else:
    print('Принять гипотезу о равенстве дисперсий')
```

```python
t, p_value  = stat.ttest_ind(x,y, axis = 0, equal_var=False)

if p_value < 0.05:
    print(f'{Ha} так как p value - {p_value.round(3)} < 0.05')
else:
    print(f'{Ho} так как p value - {p_value.round(3)} > 0.05')
```

```python
t,p_value  = stat.ttest_ind(x,y, axis = 0, equal_var=False, alternative='less')

if p_value < 0.05:
    print(f'{Ha} так как p value - {p_value.round(3)} < 0.05')
else:
    print(f'{Ho} так как p value - {p_value.round(3)} > 0.05')
```

Таким образом, получается у двух популярных сорта вина Испании и Португалии разный рейтинг, но средняя цена одинакова.
